```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import researchpy as rp
import statsmodels.api as sm
from statsmodels.formula.api import ols
    
import matplotlib.pyplot as plt
```

```python
a1 = pd.read_csv('/home/hao/Documents/QEES-lab1/evaluation/transport_time/transport_time_FastRTPS_4Kbyte.txt',sep = '\t',encoding = 'utf-8')
a2 = pd.read_csv('/home/hao/Documents/QEES-lab1/evaluation/transport_time/transport_time_OpenSplice_4Kbyte.txt',sep = '\t',encoding = 'utf-8')
a3 = pd.read_csv('/home/hao/Documents/QEES-lab1/evaluation/transport_time/transport_time_Connext_4Kbyte.txt',sep = '\t',encoding = 'utf-8')
b1 = pd.read_csv('/home/hao/Documents/QEES-lab1/evaluation/transport_time/transport_time_FastRTPS_4Mbyte.txt',sep = '\t',encoding = 'utf-8')
b2 = pd.read_csv('/home/hao/Documents/QEES-lab1/evaluation/transport_time/transport_time_OpenSplice_4Mbyte.txt',sep = '\t',encoding = 'utf-8')
b3 = pd.read_csv('/home/hao/Documents/QEES-lab1/evaluation/transport_time/transport_time_Connext_4Mbyte.txt',sep = '\t',encoding = 'utf-8')
# print(np.mean(a1)) GAO GE MAO XIAN
# a = np.concatenate((a1.T, a2.T, a3.T), axis = 1)
# b = np.concatenate((b1, b2, b3), axis = 0)
# print(a)
c = np.append(a1,a2)
data = np.append(c, a3)

c2 = np.append(b1, b2)
data2 = np.append(c2, b3)
# print(d)
```

```python
algorithm = ['FastRTPS_4KB', 'OpenSplice_4KB', 'Connext_4KB']
d = {'Group': np.repeat(algorithm,119), 'Latency': data}
df = pd.DataFrame(data=d)
# print(nont(df['Latency'])
```

```python
# Summary based on algorithm type
rp.summary_cont(df['Latency'].groupby(df['Group']))
```

```python
# stats.f_oneway(data_group1, data_group2, data_group3, data_groupN)
stats.f_oneway(df['Latency'][df['Group'] == 'FastRTPS_4KB'], 
             df['Latency'][df['Group'] == 'OpenSplice_4KB'],
             df['Latency'][df['Group'] == 'Connext_4KB'])
```

```python
# model_name = ols('outcome_variable ~ C(group_variable)', data=your_data).fit()
results = ols('Latency ~ C(Group)', data=df).fit()
results.summary()
```

```python
aov_table = sm.stats.anova_lm(results, typ=2)
aov_table
```

```python
import pandas
import researchpy as rp
import seaborn as sns
import numpy as np

import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.stats.multicomp

c = np.append(np.append(np.append(np.append(np.append(a1,a2), a3), b1), b2), b3)
print(c.shape)
df = pd.DataFrame({'Code_Size': c,
                   'Sample_size': np.repeat(['4K', '4M'], 357),
                   'Configurations': np.r_[np.repeat(['FastRTPS', 'OpenSplice', 'Connext'],119),
                                      np.repeat(['FastRTPS', 'OpenSplice', 'Connext'],119)]})
```

```python
rp.summary_cont(df.groupby(['Sample_size', 'Configurations']))['Code_Size']
```

```python
model = ols('Code_Size ~ C(Sample_size)*C(Configurations)', df).fit()

# Seeing if the overall model is significant
print(f"Overall model F({model.df_model: .0f},{model.df_resid: .0f}) = {model.fvalue: .3f}, p = {model.f_pvalue: .4f}")
```

```python
model.summary()
```

```python
res = sm.stats.anova_lm(model, typ= 2)
res
```
